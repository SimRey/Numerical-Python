## <center>Polynomials</center>
The np.polynomial module contains a number of classes for representing polynomials in different polynomial bases.
Standard polynomials, written in the usual power basis $x_i$, are represented with the Polynomial class. To create an
instance of this class, we can pass a coefficient array to its constructor. In the coefficient array, the ith element
is the coefficient of $x_i$.

```python
import numpy as np
from numpy import polynomial as P
import matplotlib.pyplot as plt
```

**Example**
Representation of the polynomial $$1+2x+3x^2$$

```python
p1 = P.Polynomial([1, 2, 3])
p1
```

```python
# To access each element:
print(p1.coef)
print(p1.domain)
print(p1.window)
```

```python
# Polynomial can be obtained form its roots
p2 = P.Polynomial.fromroots([-1, 1])  
p2
```

```python
# To calculate the roots of a polynomial use the method P.Polynomial.roots()
print(P.Polynomial.roots(p2))
```

```python
# Polynomial evaluation
p1_eval = p1(np.array([1, 1.5, 2]))  # Evaluated at 1, 1.5 and 2
print(p1_eval)
```

```python
# Instances of Polynomial can be operated on using the standard arithmetic operators +, -, *, /, and so on.
# The // operator is used for polynomial division.
# Ex:
P1 = P.Polynomial.fromroots([1, 2, 3])
P2 = P.Polynomial.fromroots([2])
P3 = P1//P2
print(P3)
print(P3.roots())
```

### Polynomial interpolation

There are many ways to perform polynomial interpolations, the best way to do it is offered by the numpy library

```python
x = np.linspace(100.,5000.,100)
y = lambda z: z + np.sin(np.pi*z/max(z))*z**2 - 1/z + np.log(z)
plt.scatter(x, y(x), label="Original data")

for i in range(1, 11):
    p = np.polyfit(x, y(x), i)
    p_val = np.polyval(p, x)

    plt.plot(x, p_val, label=f"{i}-order")

plt.legend(loc="best")
```
