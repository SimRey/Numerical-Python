### Displaying Statistical Datasets

#### a) Scatter 
This is the simplest way to represent univariate data: just plot each individual data point.

```python
plt.scatter(np.arange(len(data)), data)
plt.xlabel("Datapoints")
plt.ylabel("Values")
plt.title("Scatter plot");
```

#### b) Histograms
Histograms provide a first good overview of the distribution of your data. If you
divide by the overall number of data points, you get a relative frequency histogram;
and if you just connect the top center points of each bin, you obtain a relative
frequency polygon.

```python
sns.histplot(data=data, bins=10)
plt.xlabel("Datapoints")
plt.ylabel("Frecuencies")
```

#### c) Kernel-Density-Estimation (KDE) Plots
Histograms have the disadvantage that they are discontinuous, and that their shape
critically depends on the chosen bin-width. In order to obtain smooth probability
densities, i.e., curves describing the likelihood of finding an event in any given
interval, the technique of Kernel Density Estimation (KDE) can be used. Thereby
a normal distribution is typically used for the kernel. The width of this kernel
function determines the amount of smoothing.

The bandwidth of the kernel is the parameter which determines how much we
smooth out the contribution from each event. It can be shown that under certain conditions the optimal choice for h is:
$$ h = (\frac{4\hat{\sigma}^5}{3n})^{1/5}$$

where $\hat{\sigma}$ is the standard deviation of the samples

```python
sigma_hat = np.std(data)
sigma_hat
```

```python
bw = ((4*sigma_hat**5)/(3*len(data)))**(1/5)
bw
```

```python
sns.kdeplot(data=data, bw_adjust=bw)
plt.xlabel("Datapoints")
plt.ylabel("Frecuencies")
plt.title("KDE plot");
```

```python
bws = np.linspace(0.1, 1, 5)
for i in bws:
    sns.kdeplot(data=data, bw_adjust=i, label=f"bw: {i}")

plt.xlabel("Datapoints")
plt.ylabel("Frecuencies")
plt.title("KDE plot")
plt.legend(loc="best");
```

#### d) Cumulative Frequencies
A cumulative frequency curve indicates the number (or percent) of data with less
than a given value. This curve is very useful for statistical analysis, for example
when we want to know the data range containing 95% of all the values. Cumulative
frequencies are also useful for comparing the distribution of values in two or more
different groups of individuals.
When you use percentage points, the cumulative frequency presentation has the
additional advantage that it is bounded: $0 \leq cumfreq \leq 1$

```python
n_bins = 10
n = plt.hist(data, bins=n_bins, label="Histogram")

plt.plot(n[1], stats.cumfreq(data, (n_bins+1))[0], label="Cummulative frecuency")
plt.xlabel("Bins")
plt.ylabel("Frecuencies")
plt.title("Cumulative frequencies plot")
plt.legend(loc="best");
```

### Descriptive Statistics
Descriptive statistics can be divided into two groups:
1. Measures of central tendency
2. Measures of dispersion

##### Measures of central tendency

```python
# Mean
mean = np.mean(data)
mean
```

```python
# Median
median = np.median(data)
median
```

```python
# Mode
mode = stats.mode(data)
mode
```

##### Measures of dispersion

```python
# Range
ran = np.max(data) - np.min(data)

# Quartiles
Q1 = np.quantile(data, 0.25)
Q2 = np.quantile(data, 0.5)
Q3 = np.quantile(data, 0.75)
Q4 = np.quantile(data, 1.0)
IQR = Q3 - Q1

print(f"""Range: {ran}
Q1: {Q1}
Q2: {Q2}
Q3: {Q3}
Q4: {Q4}
IQR: {IQR}""")
```

```python
# Boxplot is usually used to visualize the previous values
plt.boxplot(x=data, vert=False)
plt.title("Boxplot")
plt.grid(alpha=0.4);
```

```python
# Percentiles
P10 = np.quantile(data, 0.1)
P90 = np.quantile(data, 0.9)

print(f"""P10: {P10}
P90: {P90}""")
```

```python
# Standard deviation, variance, skewness and kurtosis 

stand = np.std(data)
variance = stand**2

skew = stats.skew(data)
kurt = stats.kurtosis(data)

print(f"""Standard deviation: {stand}
Variance: {variance}
Skewness: {skew}
Kurtosis: {kurt}""")
```

The $R^2$ value shows that the data adjusts well to the normal probability line, showing a normal distribution
